Ways to Clean and Transform Data using MEASURES OF SPREAD



```r
#World Bank data, 2016 version
WB <- read.csv("https://raw.githubusercontent.com/jreuning/sds230_data/refs/heads/main/WB.2016.csv", header = T, as.is = T)
```

```r
#Make a boxplot - try changing horizontal = TRUE
boxplot(WB$Fertility16, 
        horizontal = TRUE,
        col = 'yellow',
        main = "Boxplot of Fertility Rate 2016", 
        xlab = "Fertility Rate 2016")
#add mean
points(mean(WB$Fertility16, na.rm = T), pch = 19, col = 'red', cex = 2) 
```

Let's compare how a boxplot and a histogram look for the same data

```r
#For comparison, let's look at a corresponding histogram
hist(WB$Fertility16, 
        col = 'yellow',
        main = "Boxplot of Fertility Rate 2016", 
        xlab = "Fertility Rate 2016")
#add line at the mean
abline(v = mean(WB$Fertility16, na.rm = T), col = 'red', lwd = 3)
abline(v = median(WB$Fertility16, na.rm = T), col = 'black', lwd = 3)
legend("topright", c("mean", "median"), col = c("red", "black"), lwd = 3)
```

Roundup Data

```r
#get data
roundup <- read.csv("https://raw.githubusercontent.com/jreuning/sds230_data/refs/heads/main/RoundUp.csv", header = T)
#look at data
roundup
```

```r
#make dataset that is just radishes
radish <- roundup[roundup$Species == "Radish", ]
table(radish$Concentration)
```

```r
#Make boxplot by concentration
boxplot(radish$Dry.Weight ~ radish$Concentration,
        col = 'red',
        main = "Radish Dry Weight vs Concentration", 
        xlab = "Round-Up Concentration", 
        ylab = "Dry Weight")
```

Variance and Standard Deviation

```r
#We look at World Bank Data
var(WB$Fertility16)
WB$Fertility16
```

```r
#again, we have to deal with missing data
var(WB$Fertility16, na.rm = T)
```

```r
#round to make it easier to read
round(var(WB$Fertility16, na.rm = T), 2)
#get standard deviations
round(sd(WB$Fertility16, na.rm = T), 2)
```

HANDY - get lots of summary statistics at once!

```r
#Summary Statistics
summary(WB$Fertility16)
summary(WB$GunsPer100)
summary(WB$CO2)
```

You can do summaries by group

```r
tapply(radish$Dry.Weight, radish$Concentration, summary)
```

```r
#get variances
round(tapply(radish$Dry.Weight, radish$Concentration, var), 3)
#get standard deviations
round(tapply(radish$Dry.Weight, radish$Concentration, sd), 3)
```
