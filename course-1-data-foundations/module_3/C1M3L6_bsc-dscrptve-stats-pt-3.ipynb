Next we get World Bank Data from 2016 to look at skewness, outliers, etc.

```r
#Get data sets
#World Bank data, 2016 version
WB <- read.csv("https://raw.githubusercontent.com/jreuning/sds230_data/refs/heads/main/WB.2016.csv", header = T, as.is = T)
#mean function fails because of missing data
mean(WB$LifeExp)
```

```r
#What happened to mean?
WB$LifeExp
```

```r
#Solution - remove missing values with na.rm = T option
mean(WB$LifeExp, na.rm = T)
round(mean(WB$LifeExp, na.rm = T), 1)
round(median(WB$LifeExp, na.rm = T), 1)
```

Next, we introduce the dotplot.  This requires that we do some rounding of life expectancy values.

```r
#As an experiment, see what happens if you don't round
stripchart(WB$LifeExp, 
           method = "stack", 
           at = 0, 
           pch = 19, 
           col = 'blue',
           main = "Dotplot of Life Expectancy", 
           xlab = "Life Expectancy")
```

Let's make next plot with mean and median added to the plot

```r
#Let's create a dotplot of GNI per capita:
stripchart(round(WB$GNI, -3.5), 
           method = "stack", 
           at = 0, 
           pch = 19,
           col = 'blue',
           main = "Dotplot of GNI per capita", 
           xlab = "GNI per Capita")
#mtext stands for margin text
#paste puts character information together
mtext(paste("Mean =", 
            round(mean(WB$GNI, na.rm = T), 0), 
            ", Median =", 
            round(median(WB$GNI, na.rm = T), 0)))
```

Example of an outlier - Guns per 100 people

```r
hist(WB$GunsPer100, 
     col = "blue", 
     main = "Histogram of Guns per 100 people", 
     xlab = "Guns per hundred", 
     ylab = "frequency")
```

```r
#identify the outlier

WB$Country[WB$GunsPer100 > 50]
#fix missing values
na.omit(WB$Country[WB$GunsPer100 > 50])
```
