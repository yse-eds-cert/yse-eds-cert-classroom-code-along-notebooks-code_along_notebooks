## Part 1: What is Data? Understanding the Data Lifecycle 📊

### What is Data?
Data is any recorded information that can be analyzed to gain insights. In environmental science, this includes:
- 📡 **Sensor readings** (temperature, pressure, chemical concentrations)
- 📸 **Images and videos** (satellite imagery, field photos, microscopy)
- 📝 **Observations** (species counts, behavior notes, field measurements)
- 🧪 **Lab results** (sample analyses, experimental outcomes)

### The Data Lifecycle
Data doesn't just appear and disappear - it follows a lifecycle:

1. **📥 Data Acquisition** → How we collect data
2. **🧹 Data Cleaning** → Making data usable
3. **⚙️ Data Processing** → Transforming and analyzing
4. **💾 Data Archiving** → Long-term preservation
5. **🌐 Data Sharing** → Making data accessible

### Why Organization Matters: The Scaling Challenge

Consider this progression:
- **Day 1**: You have 1 CSV file - easy to manage!
- **Month 1**: You have 100 files across 5 experiments
- **Year 1**: You have 10,000 files from multiple sites
- **Year 5**: You're managing millions of files, versions, and derivatives

Without proper organization:
- 🔍 Finding the right data takes hours
- 🔄 Reproducing analyses becomes impossible
- 💰 Storage costs spiral out of control
- 🤝 Collaboration turns into chaos

> "The best time to organize your data was at the start of your project. The second best time is now!"

## Part 3: FAIR Data Principles

Think of FAIR principles as the golden rules of data management. They ensure your data remains valuable long after your project ends.

### 🔍 **F**indable - Can others (and future you) find your data?
- ✅ Use descriptive names that tell a story
- ✅ Add keywords and tags
- ✅ Create a "map" (metadata) to your data treasure

### 🔓 **A**ccessible - Can people actually get to your data?
- ✅ Use common file formats (CSV, not proprietary formats)
- ✅ Provide clear instructions
- ✅ Note any restrictions (like endangered species locations)

### 🔄 **I**nteroperable - Does your data play well with others?
- ✅ Use standard terms (e.g., "temperature_celsius" not "temp")
- ✅ Follow community standards
- ✅ Include units and definitions

### ♻️ **R**eusable - Can someone use your data for new research?
- ✅ Include detailed documentation
- ✅ Specify how others can use it (licenses)
- ✅ Provide quality indicators

### Quick FAIR Check ✓
Ask yourself:
1. If I gave this data to a colleague, could they understand it in 5 minutes?
2. Will I understand this data in 5 years?
3. Could someone reproduce my analysis?

If any answer is "no", your data isn't FAIR yet!

````python
# Let's see FAIR in action with a simple example
import pandas as pd
from datetime import datetime

# ❌ NOT FAIR - What Dr. Chen had
print("❌ NOT FAIR Example:")
print("Filename: temp.csv")
print("Content: Just numbers with no context\n")

# Create a FAIR-NON-compliant data example
poor_data = pd.DataFrame({
    'date': ['Jan', '06'],
    'site_id': ['Brazil', 'Brazil2'],
    'temp': [28.5, 28.7],
    'measure1': [3.0, 3.0],
})

print(poor_data)


# ✅ FAIR - What she should have had
print("\n✅ FAIR Example:")
print("Filename: 2024-01-15_FIJI01_water_temperature_celsius_v01.csv")
print("Content: Structured data with clear headers\n")

# Create a FAIR-compliant data example
fair_data = pd.DataFrame({
    'datetime_utc': ['2024-01-15 08:00:00', '2024-01-15 09:00:00'],
    'site_id': ['FIJI01', 'FIJI01'],
    'temperature_celsius': [28.5, 28.7],
    'depth_meters': [3.0, 3.0],
    'quality_flag': ['good', 'good']  # Clear quality indicators
})

print(fair_data)

# The magic is in the clarity!
print("\n🌟 Why this is FAIR:")
print("F - Clear filename tells you what, where, when")
print("A - Simple CSV format anyone can open")
print("I - Standard units (celsius, meters) and datetime format")
print("R - Quality flags and complete information for reuse")
````

## Part 4: The 6-Folder System 📁

Think of these folders like rooms in your field station:

1. **📥 raw_data** - The vault (NEVER modify files here!)
2. **🔧 processed_data** - The workshop (cleaned and prepared data)
3. **📊 analysis** - The lab (your code and notebooks)
4. **📋 metadata** - The filing cabinet (documentation)
5. **📤 outputs** - The display room (results and figures)
6. **📦 archive** - The storage room (old versions)

### Why This Structure Works
- **Clear names** = No guessing what's inside
- **Separate raw/processed** = Protects original data
- **Dedicated metadata** = Documentation stays with data

````python
# Let's build this structure together!
import os
from pathlib import Path

def create_project_structure(project_name):
    """Create a well-organized directory structure for your research project"""
    
    # Create the main project folder
    project_path = Path(project_name)
    project_path.mkdir(exist_ok=True)
    
    # Create data and scripts directories at the project root
    data_path = project_path / "data"
    scripts_path = project_path / "scripts"
    
    data_path.mkdir(exist_ok=True)
    scripts_path.mkdir(exist_ok=True)
    
    # Define our 6-folder system inside the data directory
    data_folders = [
        "raw_data",
        "processed_data", 
        "analysis",
        "metadata",
        "outputs",
        "archive"
    ]
    
    # Create each data subfolder
    for folder in data_folders:
        folder_path = data_path / folder
        folder_path.mkdir(exist_ok=True)
        
        # Add a README to explain each folder's purpose
        readme_path = folder_path / "README.txt"
        
        # Write helpful descriptions
        if "raw_data" in folder:
            readme_path.write_text("IMPORTANT: Never modify files in this folder!\nThis is your data vault - keep originals safe here.")
        elif "processed" in folder:
            readme_path.write_text("Place cleaned and processed data here.\nAlways document what processing was done.")
        elif "analysis" in folder:
            readme_path.write_text("Extra analysis docs / notebooks go here.\nOrganize by analysis type or date.")
        elif "metadata" in folder:
            readme_path.write_text("Documentation about your data.\nInclude data dictionaries, collection methods, etc.")
        elif "outputs" in folder:
            readme_path.write_text("Final results, figures, and reports.\nReady to share with others!")
        elif "archive" in folder:
            readme_path.write_text("Old versions and superseded files.\nAlways date your archived files.")
    
    # Add README for scripts folder
    scripts_readme = scripts_path / "README.txt"
    scripts_readme.write_text("Version-controlled scripts and code.\nUse git to track changes to your analysis code.")

# Create an example project
create_project_structure("my_coral_research")
````

## Part 5: Metadata - Your Data's Story 📚 

Metadata is simply "data about your data." Think of it as leaving a detailed note for someone (including future you!) who finds your data.

### The 5 W's of Metadata 🖐️

Every dataset needs to answer:
1. **WHO** collected the data?
2. **WHAT** does the data contain?
3. **WHERE** was it collected?
4. **WHEN** was it collected?
5. **WHY** was it collected? (Plus HOW!)

### Essential Metadata Components 📝

- **Title**: Clear, descriptive name
- **Creator**: Your name and contact info
- **Date**: When data was collected
- **Location**: GPS coordinates or site names
- **Methods**: How you collected the data
- **Variables**: What each column means
- **Units**: Don't forget these!
- **Quality notes**: Any issues or limitations

### Why Metadata Matters for ML/AI 🤖

In the era of machine learning and AI, metadata is more critical than ever:

1. **Training Data Quality**: ML models need to know data provenance and quality metrics
2. **Feature Engineering**: Metadata helps identify which variables are relevant
3. **Bias Detection**: Understanding data collection methods reveals potential biases
4. **Reproducibility**: ML experiments require exact data versioning and parameters
5. **Data Discovery**: AI can automatically find and suggest relevant datasets

### Metadata Format Options 📄

Choose the right format for your needs:

| Format | Use Case | Pros | Cons |
|--------|----------|------|------|
| **.txt** | Quick notes | Simple, universal | No structure |
| **.md** | Documentation | Formatted, readable | Limited structure |
| **.json** | Automated processing | Structured, machine-readable | Less human-friendly |

Example comparison in plain text:
#### metadata.txt
```
1 Temperature data from Station A
2 Collected by Dr. Chen
3 January 2024
```
#### metadata.md
```
1 ## Temperature Data
2 **Station**: A  
3 **Collector**: Dr. Chen  
4 **Date**: January 2024  
```  
  ##### can view as markdown to get some formatting...
  ```md
1 ## Temperature Data
2 **Station**: A  
3 *Collector**: Dr. Chen  
4 **Date**: January 2024  
  ```
#### metadata.json
```
{
  "title": "Temperature Data",
  "station": "A",
  "collector": "Dr. Chen",
  "date": "2024-01"
}
```

> **Pro Tip**: Start with .md for readability, convert to .json when you need automation!

## Part 1: Understanding System Commands - Python vs R 🔧

### Why the Difference?

In Jupyter notebooks:
- **Python**: Uses `!` as a magic command to run shell commands
- **R**: Uses `system()` function with `intern = TRUE` to capture output

This difference exists because R integrates system commands as function calls, allowing better integration with R's data structures.

### Python Approach (from original lecture)

````python
# Python: Using ! for system commands
!pwd
!ls -la data/
!echo "Analyzing coral reef data"
````

### R Equivalent Implementation

````python
# R: Using system() with intern = TRUE
current_dir <- system("pwd", intern = TRUE)
print(paste("Current directory:", current_dir))

# List files
file_list <- system("ls -la data/", intern = TRUE)
cat(file_list, sep = "\n")

# Echo command
message <- system("echo 'Analyzing coral reef data'", intern = TRUE)
print(message)
````

### Key Differences Explained

| Aspect | Python `!` | R `system()` |
|--------|-----------|-------------|
| **Syntax** | `!command` | `system("command", intern = TRUE)` |
| **Output** | Prints directly | Returns as character vector |
| **Capture** | Use `%%bash` cell magic | Use `intern = TRUE` parameter |
| **Error handling** | Limited | Can use `try()` or `tryCatch()` |
| **Integration** | Separate from Python objects | Direct integration with R objects |

````python
# Base R version - No external packages needed!
create_project_structure <- function(project_name) {
  "Create a well-organized directory structure for your research project"
  
  # Create the main project folder
  dir.create(project_name, showWarnings = FALSE)
  
  # Create data and scripts directories
  data_path <- file.path(project_name, "data")
  scripts_path <- file.path(project_name, "scripts")
  
  dir.create(data_path, showWarnings = FALSE)
  dir.create(scripts_path, showWarnings = FALSE)
  
  # Define our 6-folder system
  data_folders <- c(
    "raw_data",
    "processed_data",
    "analysis",
    "metadata",
    "outputs",
    "archive"
  )
  
  # Create each data subfolder
  for (folder in data_folders) {
    folder_path <- file.path(data_path, folder)
    dir.create(folder_path, showWarnings = FALSE)
    
    # Add a README to explain each folder's purpose
    readme_path <- file.path(folder_path, "README.txt")
    
    # Write helpful descriptions
    if (grepl("raw_data", folder)) {
      writeLines(
        c("IMPORTANT: Never modify files in this folder!",
          "This is your data vault - keep originals safe here."),
        readme_path
      )
    } else if (grepl("processed", folder)) {
      writeLines(
        c("Place cleaned and processed data here.",
          "Always document what processing was done."),
        readme_path
      )
    } else if (grepl("analysis", folder)) {
      writeLines(
        c("Extra analysis docs / notebooks go here.",
          "Organize by analysis type or date."),
        readme_path
      )
    } else if (grepl("metadata", folder)) {
      writeLines(
        c("Documentation about your data.",
          "Include data dictionaries, collection methods, etc."),
        readme_path
      )
    } else if (grepl("outputs", folder)) {
      writeLines(
        c("Final results, figures, and reports.",
          "Ready to share with others!"),
        readme_path
      )
    } else if (grepl("archive", folder)) {
      writeLines(
        c("Old versions and superseded files.",
          "Always date your archived files."),
        readme_path
      )
    }
  }
  
  # Add README for scripts folder
  scripts_readme <- file.path(scripts_path, "README.txt")
  writeLines(
    c("Version-controlled scripts and code.",
      "Use git to track changes to your analysis code."),
    scripts_readme
  )
  
  # List the created structure
  cat("\nCreated folders:\n")
  all_dirs <- list.dirs(project_name, recursive = TRUE, full.names = FALSE)
  for (dir in all_dirs[all_dirs != ""]) {
    cat(sprintf("  📁 %s\n", dir))
  }
  
  return(invisible(TRUE))
}

# Create an example project
create_project_structure("my_coral_research_R")
````

## 🎯 Key Takeaways

### System Commands in R
- Use `system(command, intern = TRUE)` to capture output
- Returns results as character vectors (integrates with R)
- Can be wrapped in `try()` for error handling

### Base R for FAIR Data
- **No packages needed**: Everything works with base R
- **data.frame**: Standard tabular data structure
- **read.csv/write.csv**: Universal data exchange
- **file.path()**: Cross-platform path handling
- **dir.create()**: Directory creation

### When to Use Base R vs Packages?
- **Base R**: Maximum compatibility, no dependencies, teaching/sharing
- **Tidyverse**: Complex data manipulation, modern workflows
- **Both**: Use base R for core functionality, add packages as needed
