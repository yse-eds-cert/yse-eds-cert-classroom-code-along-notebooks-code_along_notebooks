# 🖥️ Bash/CLI Quick Start - 5 Minute Introduction

## What is Bash? 🐚

**Bash** (Bourne Again Shell) is a command-line interface that lets you control your computer with text commands instead of clicking. Think of it as texting instructions to your computer!

**Why use it?**
- Faster than clicking through folders
- Automate repetitive tasks
- Essential for remote servers
- Required for Git, data processing, and scientific computing

---

## The Filesystem: Your Computer's Filing Cabinet 📁

Your computer organizes files in a tree structure:

```
/                    (Root - top of tree)
├── home/           
│   └── sarah/      (Your home directory)
│       ├── Documents/
│       ├── Downloads/
│       └── projects/
│           └── coral_data/
│               ├── raw_data/
│               └── analysis/
```

**Key Concepts:**
- **Path**: Address to a file/folder (like `/home/sarah/projects`)
- **Current Directory**: Where you are now (your "location")
- **Parent Directory**: One level up (`..`)
- **Home Directory**: Your personal folder (`~`)

## Command Structure & Syntax Rules ⚙️  

  ### Basic Structure:  
```bash
   command [options] [arguments]
```     
```bash
    ls                  # Command only
    ls -l               # Command + option
    ls Documents        # Command + argument
    ls -la Documents    # Command + option + argument
```

  ### Complex Structure: Commands Within Commands 🔧  
```bash
    command subcommand [options] [arguments]
```
```bash
    git status              # git (command) + status (subcommand)
    git add file.txt        # git + add + argument
    git commit -m "message" # git + commit + option + argument
    git push origin main    # git + push + two arguments
```

   **Why subcommands?**
   - Groups related functionality (all Git operations start with `git`)
   - Makes complex tools more organized
   - Each subcommand can have its own options and arguments

   **Other tools with subcommands:**
   - `docker` (docker run, docker build, docker ps)
   - `npm` (npm install, npm run, npm test)
   - `conda` (conda create, conda activate, conda install)

### Critical Syntax Rules:

**1. Spaces Matter!** 🚨

````python
%%bash
# CORRECT ✅ - space between command and option
ls -l Bash*

# The following would be WRONG ❌
# ls-l        # No space between command and option
# ls -lDocuments  # No space between option and argument
````

**2. Single vs Double Dash:**
- `-` = Short option: `ls -l` (single letter)
- `--` = Long option: `ls --all` (full word)
- Can combine short: `ls -la` = `ls -l -a`

````python
%%bash
# Short option (single dash, single letter)
ls -a

# Long option (double dash, full word)
ls --all    # Same as -a

# Combining short options
# ls -la      # Same as: ls -l -a
````

**3. Comments with #:**

````python
%%bash
ls -la      # This is a comment - bash ignores everything after #
# This entire line is a comment
pwd         # Shows current directory
````

## Understanding `ls -la` Output 📊

````python
%%bash
# Create a sample directory structure to demonstrate
mkdir -p demo_folder
echo "Sample data" > demo_folder/data.csv
mkdir -p demo_folder/analysis

# Now let's look at it with ls -la
ls -la demo_folder/
````

**Output breakdown:**
```
drwxr-xr-x  5 sarah staff  160 Jan 15 14:30 .
drwxr-xr-x 12 sarah staff  384 Jan 14 09:15 ..
-rw-r--r--  1 sarah staff 2048 Jan 15 14:30 data.csv
drwxr-xr-x  3 sarah staff   96 Jan 15 13:00 analysis/
```

**Reading the columns:**
1. `drwxr-xr-x` = Permissions (d=directory, r=read, w=write, x=execute)
2. `5` = Number of links
3. `sarah` = Owner
4. `staff` = Group
5. `160` = Size in bytes
6. `Jan 15 14:30` = Last modified
7. `data.csv` = Name

**Special entries:**
- `.` = Current directory
- `..` = Parent directory
- `.hidden_file` = Files starting with `.` are hidden

````python
%%bash
# Clean up our demo
rm -r demo_folder
````

## More Useful Commands 🛠️

### File Operations

````python
%%bash
# Create a test file
echo "Line 1" > test.txt
echo "Line 2" >> test.txt
echo "Line 3" >> test.txt

# Display file contents
cat test.txt
````

````python
%%bash
# Show first 2 lines
head -n 2 test.txt

echo "---"

# Show last 2 lines
tail -n 2 test.txt
````

````python
%%bash
# Copy file
cp test.txt backup.txt

# Move/rename file
mv backup.txt test_backup.txt

# List to confirm
ls *.txt

# Clean up
rm test.txt test_backup.txt
````

### Finding Things

````python
%%bash
# Create sample files for searching
echo "The temperature is 25 degrees" > weather.txt
echo "The humidity is high today" >> weather.txt
echo "Temperature sensor reading: 23" > sensor.txt

# Search for "temperature" (case insensitive)
grep -i "temperature" *.txt
````

````python
%%bash
# Find all .txt files in current directory
find . -name "*.txt" -type f

# Clean up
rm weather.txt sensor.txt
````

### Getting Help

````python
%%bash
# Get help for ls command
ls --help | head -20  # Show first 20 lines of help
````

## Pro Tips 💡

### 1. Tab Completion
Start typing and press `Tab` to auto-complete:
```bash
cd Doc[TAB]         # Completes to: cd Documents/
```

### 2. Command History
```bash
↑ / ↓               # Navigate previous commands
history             # Show command history
!ls                 # Run last ls command
```

### 3. Wildcards

````python
%%bash
# Create sample files
touch data1.csv data2.csv data3.csv report.txt analysis.py

# List all CSV files
echo "All CSV files:"
ls *.csv

# List files matching pattern
echo "\nData files with numbers:"
ls data?.csv

# Clean up
rm data*.csv report.txt analysis.py
````

### 4. Clear Screen
```bash
clear               # Clear terminal screen (or Ctrl+L)
```

### 5. Cancel Command
```bash
Ctrl + C            # Stop running command
```

## Common Gotchas to Avoid ⚠️

1. **Case Sensitive**: `Documents` ≠ `documents`
2. **No Undo**: `rm` is permanent - no trash bin!
3. **Spaces in Names**: Use quotes or backslash
   ```bash
   cd "My Documents"     # With quotes
   cd My\ Documents      # With backslash
   ```
4. **Hidden Files**: Start with `.` (use `ls -a` to see)
