## Part 3: Working with Environmental Data Files (10 minutes)

### Adding Scientific Data and Code

#### Step 1: Create a Data Processing Script

````python
%%bash
cd ~/water-quality-analysis

# Create a Python script for data cleaning
cat > scripts/clean_sensor_data.py << 'EOF'
#!/usr/bin/env python3
"""
Clean and validate water quality sensor data.

This script handles common issues in sensor data:
- Missing values
- Sensor drift
- Out-of-range measurements
"""

import pandas as pd
import numpy as np
from datetime import datetime

def clean_sensor_data(input_file, output_file):
    """Clean raw sensor data and save processed version."""
    
    # Read raw data
    df = pd.read_csv(input_file)
    
    # Remove invalid pH values (must be 0-14)
    df.loc[(df['pH'] < 0) | (df['pH'] > 14), 'pH'] = np.nan
    
    # Remove negative dissolved oxygen
    df.loc[df['dissolved_oxygen'] < 0, 'dissolved_oxygen'] = np.nan
    
    # Flag temperature outliers
    df['temp_flag'] = (df['temperature'] < -5) | (df['temperature'] > 40)
    
    # Save cleaned data
    df.to_csv(output_file, index=False)
    print(f"Cleaned data saved to {output_file}")
    
    return df

if __name__ == "__main__":
    # Process example data
    clean_sensor_data('data/raw/sensors_2024.csv', 
                      'data/processed/sensors_2024_clean.csv')
EOF

# View the file
head -20 scripts/clean_sensor_data.py
````

### 📊 What About Large Data Files?

For files > 100 MB, consider:
1. **Git LFS** (Large File Storage)
2. **DVC** (Data Version Control)  
3. **.gitignore** to exclude them

We'll create a .gitignore file:

````python
%%bash
cd ~/water-quality-analysis

# Create .gitignore for environmental science projects
cat > .gitignore << 'EOF'

    # Large data files
    *.nc
    *.hdf5
    *.tif
    *.geotiff
    data/raw/large/

    # Temporary files
    *.tmp
    *.temp
    ~*

    # Python
    __pycache__/
    *.py[cod]
    .ipynb_checkpoints/

    # R
    .Rhistory
    .RData
    .Rproj.user/

    # System files
    .DS_Store
    Thumbs.db 
EOF

# Add and commit .gitignore
git add .gitignore
git commit -m "Add .gitignore for environmental data science projects"


# view the file in the directory
echo "
The repo directory structure is now:

$(ls -la)"


# view the contents of this file (mimics the above code)
echo "
The .gitignore file contents now: 
$(cat .gitignore)"
````

## Part 4: Branching for Experimental Analysis (10 minutes)

### Creating Branches for Different Analyses

Branches let you experiment without affecting your main analysis.

#### Scenario: Testing a New Calibration Method

````python
%%bash
cd ~/water-quality-analysis

# View current branches
git branch
echo ""

# Create and switch to new branch
git checkout -b feature/new-calibration
echo ""

# Verify you're on the new branch
git branch
````

````python
%%bash

# Note that the directory structure still looks the same
cd ~/water-quality-analysis
ls -l
````

#### Make Changes in the New Branch

````python
%%bash
cd ~/water-quality-analysis

# Create new calibration script
cat > scripts/calibrate_sensors.py << 'EOF'
#!/usr/bin/env python3
"""
New calibration method for pH sensors.
Uses temperature compensation.
"""

def calibrate_ph(raw_ph, temperature):
    """Apply temperature compensation to pH readings."""
    # Temperature compensation formula
    temp_coefficient = 0.003
    reference_temp = 25.0
    
    corrected_ph = raw_ph + temp_coefficient * (temperature - reference_temp)
    
    return round(corrected_ph, 2)

# Test the function
if __name__ == "__main__":
    test_ph = 7.5
    test_temp = 20.0
    print(f"Raw pH: {test_ph}")
    print(f"Temperature: {test_temp}°C")
    print(f"Corrected pH: {calibrate_ph(test_ph, test_temp)}")
EOF

# Add and commit in the branch
git add scripts/calibrate_sensors.py
git commit -m "Add temperature-compensated pH calibration method"
````

#### Compare Branches

````python
%%bash
cd ~/water-quality-analysis

# See what's different between branches
git diff master --name-only

# View commit history
echo ""
git log --oneline --graph --all
````

#### Merge Your Changes

````python
%%bash
cd ~/water-quality-analysis

# Switch back to main branch
git checkout master
echo ""

# Merge the feature branch
git merge feature/new-calibration
echo ""

# View updated history
git log --oneline --graph
````

# Hands-on Git Practice - Git on Windows Addendum


This addendum covers the topic of using Git on a Windows Operating System.

---

## The Windows Challenge
  Windows has fundamental differences from Unix-based systems (Mac/Linux):

  **Path Conventions**:
  - Windows: `C:\Users\sarah\data\project\file.csv`
  - Unix: `/home/sarah/data/project/file.csv`

  **Command Environment**:
  - Windows uses **Command Prompt (cmd)** or **PowerShell** - different syntax than Unix bash
  - Many scientific tools expect Unix-style commands
  - Path separators and commands work differently

## The Solution: Unix-like Environments for Windows
Rather than fighting these differences, most researchers install environments that provide Unix compatibility:  

### Option 1: MobaXterm (Lightweight Champion) 🏆  
  **Why MobaXterm?**
  - ✅ **Decade+ proven**: Trusted by researchers since 2008
  - ✅ **All-in-one**: Git, SSH, X11, and bash included
  - ✅ **Portable**: Can run from USB, no admin rights needed
  - ✅ **Easy reset**: Delete folder to completely remove/reinstall  

**Installation**:  
  1. Download from [mobaxterm.mobatek.net](https://mobaxterm.mobatek.net)
  2. Choose "Portable Edition" (no installation required)
  3. Extract and run - Git is ready immediately!

### Option 2: VS Code (Modern Powerhouse) 💪  
  **Why VS Code?**
  - ✅ **Integrated terminal**: Built-in Unix-style terminal
  - ✅ **Git built-in**: Visual Git interface + command line
  - ✅ **Extensions**: Jupyter, Python, R, and more
  - ✅ **Cross-platform**: Same experience on all OS  
  
**Installation**:
  1. Download from [code.visualstudio.com](https://code.visualstudio.com)
  2. During install, check "Add to PATH"
  3. Install "Git Bash" terminal option
  4. Git commands work in integrated terminal

### Why This Approach?

  Both options:
  - **Automatically handle path translation** (`/c/Users/` instead of `C:\Users\`)
  - **Provide bash commands** (ls, cd, grep work as expected)
  - **Include Git by default** (no separate installation)
  - **Match Mac/Linux tutorials** (commands work without translation)

  ### Quick Test

  After installation, open MobaXterm or VS Code terminal and try:
  ```bash
  git --version
  pwd            # Shows current directory (Unix style)
  ls -la         # Lists files (Unix style)
  ```
  Bottom Line

  Don't use Windows Command Prompt for Git! Install MobaXterm (simple) or VS Code (feature-rich) to get a
  consistent Unix-like experience that matches what 90% of scientific computing tutorials assume you're
  using.

  💡 Pro tip: This setup means you can follow Mac/Linux tutorials without any command translation!
