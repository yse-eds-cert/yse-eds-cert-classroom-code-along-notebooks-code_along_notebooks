## What is Version Control?

### A system that:

- 📸 **Records changes** to your data processing scripts over time
- 🔄 **Enables reverting** to when your analysis actually worked
- 👥 **Facilitates collaboration** between field teams and modelers
- 📝 **Documents why** you changed that calibration factor
- 🔍 **Tracks who** made each change to the quality control pipeline

### Think of it as:
- "Track Changes" for your entire research project
- Time machine for your code and analysis
- Lab notebook for computational work

## Key Concepts for Environmental Scientists

### Repository ("Repo")
- Your research project's organized home
- Contains all scripts, notebooks, documentation
- Like a smart folder that remembers everything

### Commit
- A snapshot of your analysis at a specific moment
- "Save point" in your research
- Example: "Fix temperature sensor calibration drift for sensors 3-5"

### Branch
- Parallel universe for testing new ideas
- Try that experimental detrending method without breaking working code
- Example: "feature/new-qc-algorithm"

### Merge
- Combining work from different branches
- Git helps identify and resolve conflicts
- Example: Integrating field team's data updates with modeling team's changes

## Git Command Line Basics 🔧

  Git follows a traditional command structure:

  git [subcommand] [flags] [arguments]

### The Structure Explained

  1. **`git`** - The main command (always starts with this)
  2. **`subcommand`** - The action you want to perform (clone, add, commit, push, AND MUCH MORE!)
  3. **`flags`** - Options that modify the behavior (start with `-` or `--`)
  4. **`arguments`** - Files, messages, or other inputs

### Essential Commands

  ```bash
  # Initialize a new repository
  git init

  # Check what's changed
  git status

  # Stage changes (prepare for commit)
  git add filename.txt        # Add specific file
  git add .                   # Add all changes

  # Commit changes (save a snapshot)
  git commit -m "Your message here"

  # View history
  git log                     # Full history
  git log --oneline          # Condensed view

  # Work with remotes
  git push                    # Send changes to GitHub
  git pull                    # Get changes from GitHub
  git clone                   # Copy a Repo from GitHub
  ```

## Security Best Practices for Environmental Data

### .gitignore:
- this is how you control what does and doesn't get added to your index nor commited to head
- protects you, protects your repository

### 🔐 Never commit:
- API keys for weather/satellite data services
- Exact GPS coordinates of sensitive sites
- Passwords or personal information
- Proprietary sensor calibration data

### ✅ Always use .gitignore for:
```gitignore
# Large data files
*.nc      # NetCDF files
*.tif     # GeoTIFF rasters
*.hdf5    # HDF5 datasets

# Sensitive information
site_coordinates_exact.csv
api_keys.txt
sensor_passwords.txt

# Temporary files
*.tmp
.ipynb_checkpoints/
__pycache__/
```

### 📊 Best practices:
- Commit metadata and processing scripts
- Store large data externally (more on this later)
- Use clear commit messages: "Fix precipitation unit conversion from mm to cm"

## 🚨 CRITICAL SECURITY WARNING: Never Commit Secrets! 🚨

  **STOP AND CHECK** before every commit:

  ### What NOT to Commit
  ❌ **API Keys** (OpenAI, AWS, Google, etc.)
  ❌ **Passwords** and authentication tokens
  ❌ **Private SSH keys** or certificates
  ❌ **Database connection strings**
  ❌ **Environment files** (.env, .env.local)

  ### Why This Matters
  - **Instant exposure**: Once pushed, assume it's compromised
  - **Permanent record**: Git history keeps deleted files forever
  - **Bot scanning**: Hackers scan GitHub for keys within seconds
  - **Real cost**: Exposed AWS keys average $10,000+ in damages

  ### Before You Commit
  ```bash
  # Always check what you're about to commit
  git status
  git diff --staged

  # If you see a secret, remove it immediately
  git reset HEAD filename

  If You Accidentally Commit Secrets

  1. IMMEDIATELY revoke/regenerate the exposed key
  2. Remove from history: git filter-branch or BFG Repo-Cleaner
  3. Force push cleaned history
  4. Assume the key is compromised - it takes <1 minute for bots to find it

## Environmental Data Versioning Challenges

### The Challenge:

| Your Scripts | Your Data |
|--------------|-----------|
| Python/R code (KB) | Satellite imagery (GB) |
| Text files | Binary files |
| Easy to diff | Impossible to diff |
| Changes frequently | Updates periodically |

### Solutions for Environmental Data:

1. **Small datasets (<100 MB)**: Include in repository
   - CSV files with sensor readings
   - Site metadata
   - Processed summary statistics

2. **Large datasets**: Use specialized approaches
   - **Git LFS**: For files up to ~2GB
   - **DVC**: Data Version Control for ML pipelines
   - **Pointers**: Store references to external data

3. **External repositories**:
   - **Zenodo**: DOIs for published data
   - **Environmental Data Initiative**: Long-term ecological data
   - **Pangaea**: Earth & environmental science data
