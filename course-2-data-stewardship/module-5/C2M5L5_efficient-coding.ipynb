## 2. How to Name Your Variables 🐉

When working with data—especially real-world environmental data—clear, descriptive variable names are essential. They make your code easier to read, debug, and share.

Let’s say we want to do some basic analysis on this dataset:

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/yse-eds-cert/yse-eds-cert-classroom-code-along-notebooks-code_along_notebooks/main/course-2-data-stewardship/module-5/example_air_quality_data.csv")
df.head()
```

The dataset includes air quality data such as temperature, humidity, PM2.5, NO₂, and ozone levels, recorded at different stations and times.

### 🚫 Bad Variable Naming

The following code works--but it's cryptic, hard to understand, and gives no hint to what the variables actually contain. This is a common trap!

```python
a = df["pm25"]
b = df["temperature_C"]
c = a.mean()
d = b.mean()
print(c, d)
```

At first glance, what do `a`, `b`, `c`, or `d` mean? You’ll forget in 10 minutes, and your collaborators will have no idea.

### ✅ Good Variable Naming

Here’s the same logic with clear, self-explanatory names:

```python
pm25_values = df["pm25"]
temp_values = df["temperature_C"]

avg_pm25 = pm25_values.mean()
avg_temp = temp_values.mean()

print("Average PM2.5:", avg_pm25)
print("Average Temperature (°C):", avg_temp)
```

This version communicates intent. Anyone reading it knows:

- What data is being used (`pm25`, `temperature_C`)
- What the variables represent (`avg_pm25`, etc.)
- That the code is calculating average air pollution and temperature

### 💡 General Variable Naming Guidelines 

**Do 👍:**
- Use lowercase with underscores (`avg_pm25`)
- Choose names that describe what the variable contains, not just how it's used
- Keep names concise but meaningful (`avg_temp` instead of `average_temperature`)
- Be consistent with naming throughout script or notebook

**Don't 👎:**
- Use spaces or special characters (`$`, `&`, ...) 
- Start with a number (`25_temp`)
- Avoid single-letter names like `l`, `O`, `I`
- Avoid reserved keywords (`class`, `for`, etc.)

## 4. Writing Effective Comments 📝

Writing comments is like leaving smart, brief notes for your future self or your teammates. But too many comments—or the wrong kind—can do more harm than good.

A good comment should:

- Explain the “why”, not restate the “what”
- Clarify intent, especially when logic isn’t obvious
- Be short, relevant, and easy to update as code changes

Think of comments like annotations in a Google Doc—you wouldn’t explain every word or punctuation mark, but you would leave a note if something might be misunderstood or needs context.

### 🚫 No Comments

```python
clean_df = df.dropna()
pm25_avg = df["pm25"].mean()
```

We don’t know why rows are dropped or what the average is used for.

### ✅ Good Comments

```python
# Drop rows where key environmental measurements are missing
clean_df = df.dropna(subset=["pm25", "temperature_C"])

# Calculate the average PM2.5 across all stations and times
pm25_avg = clean_df["pm25"].mean()
```

Now a collaborator knows:

- Which columns are essential
- Why dropna() is being used
- What kind of average is being calculated

### 🚫 Too many comments (overkill)

```python
# In this next line of code, we are going to remove all the rows in the DataFrame
# that have any null or missing values. This is important because missing data can
# affect the accuracy of our analysis, and we want to make sure we’re only working
# with clean, complete data before proceeding with further analysis steps.
clean_df = df.dropna(subset=["pm25", "temperature_C"])

# In this line, we calculate the average (or mean) value of the PM2.5 readings, which is a
# common air pollutant. We are using the 'mean()' method on the 'pm25' column of our cleaned
# DataFrame to get the overall air quality level across all sensors and time periods.
pm25_avg = clean_df["pm25"].mean()
```

Problems:

- Comments are longer than the code itself. 
- They repeat what the code already clearly expresses.
- It clutters the notebook and reduces scan-ability.
- It can quickly become outdated if the code changes but the comment doesn't.

### 💡 Best Practice for Commenting

**Comment as much as necessary, but no more.**

The best comments:

- Clarify the intent or reasoning
- Fill in what the code can’t easily say
- Stay brief, relevant, and accurate

## Summary

- ✅ Use descriptive variable names
- ✅ Organize code logically in notebooks
- ✅ Write helpful, minimal comments

## 2. How to Name Your Variables in R 🐉

R follows similar naming conventions to Python, with some R-specific considerations.

### First, let's load and explore the data in R

```python
# Load data
df <- read.csv("https://raw.githubusercontent.com/yse-eds-cert/yse-eds-cert-classroom-code-along-notebooks-code_along_notebooks/main/course-2-data-stewardship/module-5/example_air_quality_data.csv")

# Display first few rows
head(df)

# Check structure
str(df)
```

### 🚫 Bad Variable Naming in R

```python
# Bad naming - cryptic and unclear
a <- df$pm25
b <- df$temperature_C
c <- mean(a, na.rm = TRUE)
d <- mean(b, na.rm = TRUE)
print(c(c, d))
```

### ✅ Good Variable Naming in R

```python
# Good naming - clear and descriptive
pm25_values <- df$pm25
temp_values <- df$temperature_C

avg_pm25 <- mean(pm25_values, na.rm = TRUE)
avg_temp <- mean(temp_values, na.rm = TRUE)

cat("Average PM2.5:", avg_pm25, "\n")
cat("Average Temperature (°C):", avg_temp, "\n")
```

### 💡 R-Specific Naming Conventions

**R Style Choices** (all are valid):
- `snake_case`: `avg_pm25` (recommended, matches Python)
- `camelCase`: `avgPm25` (common in some R packages)
- `dot.case`: `avg.pm25` (older R style, avoid for new code)

**R-Specific Don'ts**:
- Don't use `.` at the start (`.temp` means hidden object)
- Avoid `T` and `F` as names (they're shortcuts for TRUE/FALSE)
- Don't use function names (`c`, `mean`, `data`, etc.)

## 3. Organizing Code in R Notebooks 🧱

Let's organize the same analysis workflow in R:

### A. Load Libraries (Base R - No packages needed!)

```python
# For base R, no library loading needed!
# But if using packages, you would do:
# library(stats)  # Already loaded by default

# Check R version
cat("R version:", R.version.string, "\n")
```

### B. Import Data

```python
# Read CSV file
air_quality_df <- read.csv("https://raw.githubusercontent.com/yse-eds-cert/yse-eds-cert-classroom-code-along-notebooks-code_along_notebooks/main/course-2-data-stewardship/module-5/example_air_quality_data.csv", 
                           stringsAsFactors = FALSE)

# Check dimensions
cat("Data dimensions:", nrow(air_quality_df), "rows x", 
    ncol(air_quality_df), "columns\n")
```

### C. Clean and Prepare Data

```python
# Remove rows where key values are missing
# In base R, we use complete.cases() or manual subsetting
clean_df <- air_quality_df[
  !is.na(air_quality_df$temperature_C) & !is.na(air_quality_df$pm25), 
]

# Convert timestamp to Date/POSIXct format
clean_df$timestamp <- as.POSIXct(clean_df$timestamp, 
                                 format = "%m/%d/%y %H:%M")

# Report cleaning results
rows_removed <- nrow(air_quality_df) - nrow(clean_df)
cat("Rows removed:", rows_removed, "\n")
cat("Clean data rows:", nrow(clean_df), "\n")
```

### D. Explore Data

```python
# Summary statistics in base R
summary(clean_df[, c("temperature_C", "pm25", "no2_ppb", "ozone_ppb")])

# Check for duplicate timestamps per station
duplicates <- duplicated(clean_df[, c("timestamp", "location")])
cat("\nDuplicate rows:", sum(duplicates), "\n")

# Calculate means by location using aggregate()
location_means <- aggregate(
  cbind(pm25, temperature_C) ~ location, 
  data = clean_df, 
  FUN = mean, 
  na.rm = TRUE
)

cat("\nAverage values by location:\n")
print(location_means)
```
